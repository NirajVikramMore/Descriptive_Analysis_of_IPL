# Creating New Columns

# Grouping and Aggregating

```python
ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean()
```

```python
soldprice_by_age = ipl_auction_df.groupby('AGE')['SOLD PRICE'].mean().reset_index()
print(soldprice_by_age)
```

```python
sns.barplot(x = 'AGE', y = 'SOLD PRICE', data = soldprice_by_age);
```

```python
country_counts = ipl_auction_df.COUNTRY.value_counts()
fig = px.pie(values=country_counts.values, names=country_counts.index, title='Distribution of Players by Country')
fig.show()
```

```python


import matplotlib.pyplot as plt
plt.hist(ipl_auction_df['SOLD PRICE'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Sold Price')
plt.ylabel('Frequency')
plt.title('Distribution of Sold Prices')
plt.show()
```
